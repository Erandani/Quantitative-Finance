```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import math
```

```python
# Define the ticker symbols for the indexes you want
tickers = ['PG', 'BEI.DE']  # S&P 500, Dow Jones, Nasdaq, Russell 2000

# Download the historical data for the indexes
data = yf.download(tickers, start='2020-01-01', end='2025-01-01')['Close']
```

```python
df = data.reset_index()
df = pd.DataFrame(df.values)
df.columns = [	'Date', 'PG', 'BEI.DE' ]
df = df.set_index('Date')
```

```python
df.head()
```

```python
df['PG'] = pd.to_numeric(df['PG'])
df['BEI.DE'] = pd.to_numeric(df['BEI.DE'])
sec_returns=np.log(df/df.shift(1))
```

Calculating Portfolio Risk

```python
## equal weight scheme
weights = np.array([0.5,0.5])
```

Portfolio Variance

```python
port_var = np.dot(weights.T, np.dot(sec_returns.cov()*250, weights))
port_var
```

Portfolio Volatility

```python
port_vol = (np.dot(weights.T, np.dot(sec_returns.cov()*250, weights)))**0.5
```
